```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('uber-lyft-cab-prices\cab_rides.csv')
count = sns.countplot(x='cab_type', hue='source', data = data)
```

```python
plt.figure(figsize=(16, 10))
dist_prices = sns.scatterplot(x='distance', y='price', hue='cab_type', data=data)
```

```python
plt.figure(figsize=(16,10))
hist_dist_prices = sns.catplot(x='distance', y='price', hue='cab_type', kind='bar', data=data)
```

```python
plt.figure(figsize=(16, 10))
source_prices = sns.violinplot("source", "price", data=data)
source_prices.set_xticklabels(source_prices.get_xticklabels(), rotation=45)
```
