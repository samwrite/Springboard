```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

user_engagement_df = pd.read_csv('takehome_user_engagement.csv')
user_engagement_df.head(10)
```

```python
user_df = pd.read_csv('takehome_users.csv')
user_df.head()
```

```python
adopted_list = []
```

```python
# def adopted(df):
start_index = 0
for index, row in user_engagement_df.iterrows():
    if index - start_index == 2:
        if user_engagement_df.iloc[index]['user_id'] != user_engagement_df.iloc[start_index]['user_id']:
            start_index = index - 1
        else:
            if pd.to_datetime(user_engagement_df.iloc[index]['time_stamp']) - pd.to_datetime(user_engagement_df.iloc[start_index]['time_stamp']) <= pd.Timedelta('7 days'):
                adopted_list.append(user_engagement_df.iloc[index]['user_id'])
            start_index = index - 1
adopted_list
```

```python
mylist = list(set(adopted_list))
```

```python
mylist.sort()
mylist
```

```python
d = {'object_id': mylist, 'adopted': 1}
adopted_df = pd.DataFrame(data=d)
adopted_df.head()
```

```python
user_df.set_index('object_id', inplace=True)
user_df.head()
```

```python
adopted_df.set_index('object_id', inplace=True)
adopted_df.head()
```

```python
joined_df = user_df.join(adopted_df, how='left')
joined_df.head()
```

```python
joined_df['adopted'] = joined_df['adopted'].fillna(0)
joined_df.head()
```

```python
def get_email(email):
    return email.split("@")[1]
```

```python
joined_df['email'] = joined_df.email.apply(get_email)
joined_df.head()
```

# Feature importance

remove name since its a categorical variable which is unimportant in this case.

```python
del joined_df['name']
del joined_df['last_session_creation_time']
del joined_df['creation_time']
joined_df.head()
```

```python
joined_df = joined_df.fillna(0)
```

```python
from sklearn.model_selection import train_test_split

dummied_data = pd.get_dummies(joined_df, columns=['email', 'creation_source'], drop_first=True)
y = joined_df['adopted']
X = dummied_data.drop(columns = ['adopted'])
feature_list = list(X.columns)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=10, random_state = 42)

#train model
rf.fit(X_train, y_train)
```

```python
predictions = rf.predict(X_test)

#calculate absolute errors
errors = abs(predictions - y_test)
print('Mean absolute error: ', round(np.mean(errors), 2), 'degrees.')
print(predictions)
```

```python
#calculate mean absolute percentage error
mape = 100 * (errors / y_test)
accuracy = 100 - np.mean(mape)
print('Accuracy: ', round(accuracy, 2), '%.')
```

```python
importances = list(rf.feature_importances_)

feature_importances = [ (feature, round(importance, 5)) for feature, importance in zip(feature_list, importances)]

# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```
